```python
import pandas as pd
data=pd.read_csv('metadata.csv')
# Load with relaxed settings
data = pd.read_csv(
    "metadata.csv",     # preventing dtype guessing issues
    dtype=str,            # force everything as string first (safe for cleaning)
    on_bad_lines="skip",  # skip broken rows
    engine="python"       # more tolerant parser
)
print(data.head(10))
```

```python
print(data)
```

```python
# Drop columns that are mostly empty (e.g. >95% NaN)
thresh = len(data) * 0.05   # keep only if at least 5% non-null
data = data.dropna(axis=1, thresh=thresh)

print("After dropping empty columns:", data.shape)

data["publish_time"] = pd.to_datetime(data["publish_time"], errors="coerce") #time formating
data = data.drop_duplicates(subset=["title", "abstract"], keep="first") #removing duplicates

# Fill missing text with empty string
for col in ["title", "abstract", "authors", "journal"]:
    if col in data.columns:
        data[col] = data[col].fillna("").str.replace("\n", " ").str.strip()

```

```python
print(data.columns)
print(data.isna().sum())
```

```python
# Extracting the  years
data['year'] = data['publish_time'].dt.year


#counting papers by publication year
papers_by_year = data['year'].value_counts().sort_index()
print(papers_by_year)
```

```python
#top journals publishing
top_journals = data['journal'].value_counts().head(10)
print(top_journals)
```

```python
#finding the most frequent words in titles
from collections import Counter
import re

# Join all titles into one string
all_titles = " ".join(data['title'].dropna())

# Simple tokenization (split by non-letters, lowercase)
words = re.findall(r'\b[a-zA-Z]{3,}\b', all_titles.lower())

# Count frequency
word_freq = Counter(words).most_common(20)
print(word_freq)
```

```python
#ploting number of publications overtime

papers_by_year.plot(kind="line", marker="o")
plt.title("Number of COVID-19 Publications Over Time")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
top_journals.plot(kind="bar", color="skyblue")
plt.title("Top Journals Publishing COVID-19 Research")
plt.xlabel("Journal")
plt.ylabel("Number of Papers")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
#word cloud of paper titles
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(all_titles)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Paper Titles")
plt.show()
```

```python
sns.histplot(data['journal'].value_counts(), bins=30, kde=False)
plt.title("Distribution of Paper Counts by Journal")
plt.xlabel("Number of Papers")
plt.ylabel("Frequency (Journals)")
plt.show()
```
